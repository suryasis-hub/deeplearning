```python
df = pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv')
df.head()
```

Loading the dataset and seeing the results. 

```python
import matplotlib.pyplot as plt
groups = df.groupby('Outcome')
for name, group in groups:
    plt.plot(group.Age, group.Glucose, marker='o', linestyle='', markersize=12, label=name)
```

Looks like age is not a factor, however glucose is a high predictor.


```python
for name, group in groups:
    plt.plot(group.Insulin, group.BMI, marker='o', linestyle='', markersize=12, label=name)
plt.xlabel('Insulin')
plt.ylabel('BMI')
plt.show()
```

```python
groups = df.groupby('Outcome')
for name, group in groups:
    plt.plot(group.DiabetesPedigreeFunction, group.BloodPressure, marker='o', linestyle='', markersize=8, label=name)
plt.xlabel('Diabetes Pedigree Function')
plt.ylabel('Blood Pressure')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

training_data, testing_data = train_test_split(df, test_size=0.2)
print(training_data.shape)
print(testing_data.shape)
```

Split the data as required as demanded for testing.


```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
y_train = training_data['Outcome']
X_train =  training_data.iloc[:,:-1]
knn.fit(X_train, y_train)
```

Loading Knn with n_neighbors = 1 

```python
X_test = testing_data.iloc[:,:-1]
y_test = testing_data['Outcome']
pred = knn.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, pred))  
print(classification_report(y_test, pred))
```

```python
error_rate = []
for i in range(1, 200):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
print(error_rate)
```

```python
plt.scatter(range(1,200), error_rate)
plt.xlabel('K')
plt.ylabel('Error')
plt.show()
```

```python
minpos = error_rate.index(min(error_rate))
print(minpos + 1)
print(min(error_rate))
```

```python
knn = KNeighborsClassifier(n_neighbors =  minpos + 1)
knn.fit(X_train,y_train)  
knn.score(X_test, y_test)
```
