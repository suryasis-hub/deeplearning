```python
education_csv = pd.read_csv("Desktop/deeplearning/StatewiseElementary.csv")
education_csv.head(40)
```

Now this data set has the educational details also the population. Though we do not know the units but does not matter.

```python
#AC_YEAR ,STATCD, STATNAME, DISTRICTS, BLOCKS, VILLAGES,CLUSTERS,TOTPOPULAT,P_URB_POP,POPULATION_0_6,GROWTHRATE ,SEXRATIO,P_SC_POP,P_ST_POP,OVERALL_LI,FEMALE_LIT,MALE_LIT,AREA_SQKM
statewise_relevant_csv = education_csv[["STATNAME","TOTPOPULAT","SEXRATIO","OVERALL_LI","FEMALE_LIT"]]

statewise_relevant_csv.head()
```

```python
statewise_relevant_csv.shape
```

```python
 rape_victims_csv.shape
```

```python
statewise_relevant_csv = statewise_relevant_csv.rename(columns={"STATNAME":"State/UT"})
statewise_relevant_csv.head()
```

```python
stateNamesFromEducation = list(statewise_relevant_csv['State/UT'])
for state in stateNamesFromEducation:
    print(state)
```

```python
listOfStates = rape_victims_csv.index
for state in listOfStates:
    print(state)
```

```python
capitalizedListOfStates = [];
for state in listOfStates:
    capitalizedListOfStates.append(state.upper())
print(capitalizedListOfStates)
```

```python
diffList =  list(set(stateNamesFromEducation) - set(capitalizedListOfStates)) + list(set(capitalizedListOfStates) - set(stateNamesFromEducation))
print(diffList)
```

```python
capitalizedListOfStates = [sub.replace('DELHI UT', 'DELHI') for sub in capitalizedListOfStates]
capitalizedListOfStates = [sub.replace('D&N HAVELI', 'DADRA & NAGAR HAVELI') for sub in capitalizedListOfStates]
diffList =  list(set(stateNamesFromEducation) - set(capitalizedListOfStates)) + list(set(capitalizedListOfStates) - set(stateNamesFromEducation))
print(diffList)
```

```python
for i in range(0,len(listOfStates)):
    print(listOfStates[i] + " " + capitalizedListOfStates[i])
```

```python
rape_victims_csv.head(36)
```

```python
rape_victims_csv.index = capitalizedListOfStates
rape_victims_csv.head(36)
```

```python
statewise_relevant_csv.head()
```

```python
statewise_relevant_csv.set_index('State/UT',inplace=True)
statewise_relevant_csv.head()
```

```python
combined_csv = statewise_relevant_csv.join(rape_victims_csv)
```

```python
combined_csv.head(36)
```

```python
combined_csv['VictimsPerPop'] = combined_csv['Total Victims']/combined_csv['TOTPOPULAT']
```

```python
combined_csv.head(36)
```

```python
largestDf = combined_csv.nlargest(3, ['VictimsPerPop'])
plt.bar(largestDf.index,largestDf['VictimsPerPop']*100, width = 0.3)
plt.xlabel('States')
plt.ylabel('Victims')
```

```python
combined_csv['VictimsPerPop'].corr(combined_csv['OVERALL_LI'])
```

```python
plt.scatter(combined_csv['OVERALL_LI'],combined_csv['VictimsPerPop']*100)
plt.xlabel("Overall literacy")
plt.ylabel("Rape Victims per population")
plt.show()
```

Not a great correlation on overall literacy rate and number of rape cases.However there is a negative correlation.

```python
combined_csv['VictimsPerPop'].corr(combined_csv['FEMALE_LIT'])
```

```python
plt.scatter(combined_csv['OVERALL_LI'],combined_csv['VictimsPerPop']*100 )
plt.xlabel("Female literacy")
plt.ylabel("Rape Victims per population")
plt.show()
```

Again not a great correlation on Female literacy rate and number of rape cases. However there is a negative correlation.

```python
combined_csv['CasesReportedPerPop'] =  combined_csv['Cases Reported']/combined_csv['TOTPOPULAT']
combined_csv.head()
```

```python
combined_csv['CasesReportedPerPop'].corr(combined_csv['VictimsPerPop'])
```

Since cases reported per population has strong correlation with casesreportedperpop, these are not two independent variables hence we have no way of identifying from this data that what effect more education has on case reporting.

```python
combined_csv['SEXRATIO'].corr(combined_csv['VictimsPerPop'])
```

```python
plt.scatter(combined_csv['SEXRATIO'],combined_csv['VictimsPerPop'])
plt.xlabel("Sex Ratio")
plt.ylabel("Rape Victims per population")
plt.show()
```

More presence of women per capita does not solve the rape problem.
